```python
print("When users fill out the form and receive their results, what will your custom caption be?")
print("An example would be: 'Your result is' or 'Your predicted diagnosis will be'")
prediction_label = input("What would you like your prediction text to be?")
print("Great!  Here's your app.py file.")
print("Make sure to copy paste this into the app.py file!")

app_py_code = '''import numpy as np
from flask import Flask, request, jsonify, render_template
import pickle

app = Flask(__name__) #Initialize the flask App
model = pickle.load(open('model.pkl', 'rb'))

@app.route('/')
def home():
    return render_template('index.html')

@app.route('/predict',methods=['POST'])
def predict():
    int_features = [x for x in request.form.values()]
    final_features = [np.array(int_features)]
    prediction = model.predict(final_features)

    output = round(prediction[0], 2)

    return render_template('index.html', prediction_text='''+ "'" + prediction_label + " {}" + "''" + '''.format(output))

if __name__ == "__main__":
    app.run(debug=True)
'''
```
